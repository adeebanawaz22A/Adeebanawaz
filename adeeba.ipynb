```python
def location(row):
    wtype_seq = row["WildType.Sequence"]
    mutant_seq = row["Mutant.Sequence"]
    mutant_loc = [i for i in range(min(len(wtype_seq), len(mutant_seq))) if wtype_seq[i] != mutant_seq[i]]
    if any(position < 1000 for position in mutant_loc):
        return "promoter"
    elif any(position >= 1000 for position in mutant_loc):
        return "CDS"
    else:
        return "no mutation"
new_dt_list["mut_position"] = new_dt_list.apply(location, axis=1)
```

```python
# Define a function to categorize the cell viability effect
def CellViabilityEffect(row):
    if row['CellViability.Mut.Rep_mean'] > 0:
        return 'Increase'
    elif row['CellViability.Mut.Rep_mean'] < 0:
        return 'Decrease'
    else:
        return 'No Effect'

# Apply the function to create a new column 'CellViabilityEffect'
new_dt_list['CellViabilityEffect'] = new_dt_list.apply(CellViabilityEffect, axis=1)

# Sort genes based on cell viability effect
sorted_genes = new_dt_list.sort_values(by='CellViabilityEffect', ascending=False)

# Get the top 5 genes
top5_genes = sorted_genes.head(5)

# Print the top 5 genes along with relevant columns
print(top5_genes[['Gene', 'mt_type', 'CellViabilityEffect', 'mut_position', 'CellViability.Mut.Rep_mean']])




```

```python
import matplotlib.pyplot as plt
import seaborn as sns

fig, axes = plt.subplots(3, 1, figsize=(5, 10))

sns.boxplot(x='Gene', y='CellViability.Mut.Rep_mean', hue='mt_type', data=top5_genes, palette='Set2', ax=axes[0])
axes[0].set_ylabel('CellViability.Mut.Rep_mean')

sns.scatterplot(x='Gene', y='CellViability.Mut.Rep_mean', hue='mt_type', data=top5_genes, palette='Set2', ax=axes[1])
axes[1].set_xlabel('Gene')
axes[1].set_ylabel('CellViability.Mut.Rep_mean')

sns.kdeplot(data=top5_genes, x='CellViability.Mut.Rep_mean', hue='mt_type', palette='Set2', ax=axes[2])
axes[2].set_xlabel('CellViability.Mut.Rep_mean')
axes[2].set_ylabel('Density')

plt.tight_layout()
plt.show()

```
